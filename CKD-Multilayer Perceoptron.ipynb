```python
# Function to split the dataset
# Seperating the target variable
X = balance_data.values[:, 0:24]
Y = balance_data.values[:, -1]

# Spliting the dataset into train and test
X_train, X_test, y_train, y_test = train_test_split( 
X, Y, test_size = 0.1, random_state = 0)
     
# Training the Neural Network 
Deep_neural_net=MLPClassifier(hidden_layer_sizes=(100, 100, 100, 100, 100), activation='relu',learning_rate_init=0.01, solver='adam',verbose=True)
Deep_neural_net.fit(X_train, y_train)
```

```python
# Making Predictions
y_pred = Deep_neural_net.predict(X_test)

# Calculating accuracy
print("Results:")
print("Confusion Matrix: \n",confusion_matrix(y_test,y_pred))
print ("Accuracy : ",accuracy_score(y_test,y_pred)*100)
print("Report : \n",classification_report(y_test, y_pred))
```
